```python
import requests
from bs4 import BeautifulSoup    # Import Libraries
import pandas as pd
import string

response=requests.get("https://www.youtube.com/")
html_data=response.content
soup=BeautifulSoup(html_data, "html.parser")     # Scrapping www.youtube.com


    
```

```python

links=[]
watch=[]
video_links=[]
video_likes=[]
video_dislikes=[]
video_views=[]
video_types=[]                      # Lists to store the data in the DataFrame
video_titles=[]
video_images=[]
video_channel_names=[]
video_subscribers=[]


likes_video=""
dislikes_video=""
subscribers=""
channel_name=""               # Variables to store data of each link
total_views=""
video_title=""
video_type=""

total=0
for video_link in soup.find_all("a"):
    links.append(video_link)                        # Find link of the videos.

for video_link in links:
    watch_link=str(video_link.get("href"))
    if watch_link.startswith("/watch?v="):
        watch_link="https://youtube.com"+watch_link          # The link starts with "/watch" so we have to append the link with "hhtps://www.youtube.com
        if watch_link not in watch:
            watch.append(watch_link)
            total+=1

print("Total Videos: ", total)

watch=set(watch)      # To remove duplicate links

for video_link in watch:
    print(video_link)
    video_links.append(video_link)
    response=requests.get(video_link)
    html_data=response.content
    soup=BeautifulSoup(html_data,"html.parser")

    for views_var in soup.find_all("div",{"class":"watch-view-count"}):
        total_views = views_var.next
        print("Video Views", total_views)               # Extract the Total Views
        video_views.append(total_views)                     
        
    check_run=0
    
    for likes in soup.find_all("button"):
        like_count=likes.get("aria-label")
        if str(like_count).startswith("like"):
            likes_video=str(str(like_count).split()[5])
            if likes_video not in video_likes:                  # Extract the Total Likes
                print("Likes", likes_video)
                video_likes.append(likes_video)
                
        if str(like_count).startswith("dislike"):
            dislikes_video=str(str(like_count).split()[5])
            if dislikes_video not in video_dislikes:
                print("dislikes", dislikes_video)                     # Extract the Total Dislikes
                video_dislikes.append(dislikes_video)
                check_run=1
                
    if check_run==0:
        print("Likes NA")
        print("Dislikes NA")
        video_dislikes.append("NA")
        video_likes.append("NA")
        
            
    for type_find in soup.find_all("meta"):
        video_type=type_find.get("content")
    print("video type", video_type)
    video_types.append(video_type)
        
    for video_title in soup.find_all("title"):
        video_title=str(video_title)
        print("title", video_title[7:-18])                 # Extract the Video Title
        video_titles.append(video_title[7:-18])
        
    for video_img in soup.find_all("link"):
        if str(video_img).startswith("<link href=\"https://i.ytimg.com/vi"):
            video_img=str(video_img)
            video_img=video_img[12:]
            video_img=video_img.split(" ")[0]
            length=len(video_img)
            video_img=video_img[0:length-1]                   # Image Link
            print(video_img)
            video_images.append(video_img)
            break
            
    for channel in soup.find_all("div", {"class":"unsubscribe-confirmation-message"}):
        channel_name=str(channel.next)[26:]
        length=len(channel_name)
        channel_name=str(channel_name)                                # Channel Name
        channel_name=channel_name.replace("?","")
        print("Channel Name", channel_name)
        video_channel_names.append(channel_name)
    
    check_run=0
    
    for subscriber in soup.find_all("span"):
        subscriber_number=subscriber.get("aria-label")
        if subscriber_number is not None and subscriber_number!="Unsubscribe" and subscriber_number!="Subscribe" and subscriber_number!="Verified":
            check_run=1
            print("Subscribers",subscriber_number)
            video_subscribers.append(subscriber_number)            # Total Subscribers
            
    if check_run==0:
        print("Subscribers NA")
        video_subscribers.append("NA")
        
    
     # Create DataFrame
                            
df=pd.DataFrame({"Video Title":video_titles,"Likes":video_likes,"Dislikes":video_dislikes,"Views":video_views, "Video Type":video_types,"Video Image Link":video_images, "Video Link":video_links, "Channel Name": video_channel_names, "Subscribers":video_subscribers})

df.to_csv("You Tube Data")
```
